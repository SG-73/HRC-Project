#### 1. Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions

```python
plt.figure(figsize=(15,7))
df.boxplot('amount_in_usd')
```

```python
def check(x, up, lo, up_lt, lo_lt):
    if x>up_lt or x<lo_lt:
        return up
    elif x<lo_lt:
        return lo
    else:
        return x
```

```python
def outlier(dataset, column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3-Q1
    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR
    return dataset.apply(lambda row: check(row[column], Q3, Q1, upper_limit, lower_limit), axis=1)
```

```python
df['amount_in_usd_median'] = outlier(df, 'amount_in_usd')
```

### Creating a function to convert Delivery day to delivery type (1->fastest, ..., 7->slowest)

```python
def f1(x):
    if x>=0 and x<5:
        return 1
    elif x>=5 and x<10:
        return 2
    elif x>=10 and x<15:
        return 3
    elif x>=15 and x<20:
        return 4
    elif x>=20 and x<25:
        return 5
    elif x>=25 and x<30:
        return 6
    else:
        return 7
```

```python
df['DEL_TYPE'] = df.apply(lambda row: f1(row['DELIVERY_DAY_med']), axis=1)
```

### Datetime manipulation

```python
def difference_in_days(melt, ffday, lags, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-' + str(i) + 'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    return melt
```

```python
df_temp = difference_in_days(df.drop(['PURCHASE_ORDER_TYPE'], axis=1), 1, 7, 'CUSTOMER_NUMBER', 'ORDER_CREATION_DATE', 'amount_in_usd_median')
for i in range(1, 8):
    df_temp['Last-' + str(i) + 'day_Sales'] = df_temp['Last-' + str(i) + 'day_Sales'].fillna(0)
```

```python
df_temp['RELEASED_CREDIT_VALUE'] = df_temp['RELEASED_CREDIT_VALUE'].replace(to_replace='-', value='', regex=True)
df_temp['RELEASED_CREDIT_VALUE'] = df_temp['RELEASED_CREDIT_VALUE'].replace(to_replace=',', value='.', regex=True)
df_temp['RELEASED_CREDIT_VALUE'] = df_temp['RELEASED_CREDIT_VALUE'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Linear Regression

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from sklearn.linear_model import LinearRegression

# Initialize and train the regression model
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

# Make predictions on the test data
linear_regression_pred = linear_regression.predict(X_test)

# Evaluate the model
lin_mse = mean_squared_error(y_test, linear_regression_pred)
lin_rmse = mean_squared_error(y_test, linear_regression_pred, squared=False)
lin_mae = mean_absolute_error(y_test, linear_regression_pred)
lin_r2_test = r2_score(y_test, linear_regression_pred)
lin_r2_train = linear_regression.score(X_train, y_train)

print("Linear Regression: ")

print("MSE:", lin_mse)
print("RMSE:", lin_rmse)
print("MAE:", lin_mae)
print("R2_test:", lin_r2_test)
print("R2_train:", lin_r2_train)
```

### Decision Tree

```python
from sklearn import tree

# Initialize and train the regression model
dtree = tree.DecisionTreeRegressor()
dtree.fit(X_train, y_train)

# Make predictions on the test data
dtree_pred = dtree.predict(X_test)

# Evaluate the model
dt_mse = mean_squared_error(y_test, dtree_pred)
dt_rmse = mean_squared_error(y_test, dtree_pred, squared=False)
dt_mae = mean_absolute_error(y_test, dtree_pred)
dt_r2_test = r2_score(y_test, dtree_pred)
dt_r2_train = dtree.score(X_train, y_train)

print("Decision Tree Regressor: ")

print("MSE:", dt_mse)
print("RMSE:", dt_rmse)
print("MAE:", dt_mae)
print("R2_test:", dt_r2_test)
print("R2_train:", dt_r2_train)
```
